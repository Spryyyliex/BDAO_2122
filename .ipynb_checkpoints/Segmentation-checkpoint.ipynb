# Page Segmentation

This code will segment the pages on your website on your site. Basically we are grouping pages together in categories dependent on their URL structure. We can then start to build a picture of our website. This is very helpful on older sites.<br>
The quality of output will depend on how well the website has been set up! For example, run the code on p-tech.org.uk ðŸ˜†.<br><br>
This example will use one of my favourite eCommerce websites: https://shop.pimoroni.com/. <br>
You need to do a Screaming Frog crawl and export 'Internal All' as csv. It will save as internal_all.csv.<br>
This can be done with the free version of Screaming Frog.<br>
The internal_all_pimoroni.csv can be downloaded from Moodle > Resources.<br>

Based on code from @rvth

```python
# Importing all needed libraries
import pandas as pd
import io
import re
```

```python
# insert crawl csv
data = ('/Users/markbonnett/Downloads/internal_all_pimoroni.csv') 

# update to include columns you would like to include from csv. We'll just need the url data.
columns = ['Address']
df = pd.read_csv(data , usecols=columns)
```

```python
df.head
```

```python
# displaying the DataFrame
display(df)
```

```python
# The manual approach
# Here you can define your own custom regex segments

segment_definitions = [
    [(r'\/products\/'), 'products'],
    [(r'\/collections\/'), 'collections'], # keep adding segments as you need
    ]

use_segment_definitions = True

def get_segment(Address):
    
    if use_segment_definitions == True:
        for segment_definition in segment_definitions:
            if re.findall(segment_definition[0], Address):
                return segment_definition[1]
        return 'Other'

# Add a segment column, and assign each URL a segment   
df['segment'] = df['Address'].apply(lambda x: get_segment(x))
```

```python
df.head(10)
```

```python
# Visualise the output
df['segment'].value_counts().plot.bar()
```

```python
# We can automated it.
# Function to create a segment from a URL
def get_segment(Address):
  
        slug = re.search(r'https?:\/\/.*?\//?([^\/]*)\/', Address)
        if slug:
            return slug.group(1)
        else:
            return 'None'

# Add a segment column, and assign each URL a segment   
df['segment'] = df['Address'].apply(lambda x: get_segment(x))
```
